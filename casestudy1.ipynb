```python
#4.Find the top 5 cars based on their sales price.
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
data=pd.read_csv('/content/sample_data/cars_data.csv')
price_df=data.groupby('Make')['Sale Price'].max().reset_index()
plt.figure(figsize=(10,10))
plt.hist(price_df['Make'],rwidth=0.8)
plt.xticks(rotation=90)
```

```python
#Find the least 5 cars based on their Resell price
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
data=pd.read_csv('/content/sample_data/cars_data.csv')
price_df=data.groupby('Make')['Resell Price'].min().reset_index()
plt.figure(figsize=(10,10))
plt.hist(price_df['Make'],rwidth=0.9)
plt.xticks(rotation=90)
```
